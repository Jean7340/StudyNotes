```python
from statsmodels.tsa.stattools import grangercausalitytests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("AdSpend_Sales.csv")
data = data.set_index('Week')
data
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data.index, data["Ad_Spend"], label="Ad Spend", alpha=0.8)
plt.plot(data.index, data["Sales"], label="Sales", alpha=0.8)
plt.title("Time Series: Ad Spend vs. Sales")
plt.xlabel("Week")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
# Perform Granger Causality test
test_result = grangercausalitytests(data[["Sales", "Ad_Spend"]], maxlag = 5)

wald_stat = test_result[1][0]['ssr_ftest'][0]
p_value = test_result[1][0]['ssr_ftest'][1]
print(f"Lag {1}: Wald statistic = {wald_stat}, p-value = {p_value}")
```
