```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
# Load and visualize a toy dataset
df = pd.read_csv('https://drive.google.com/uc?id=1X5Pxu_RCsYnFbKLVLt3QOzACR0TStGrI&export=download', index_col=0)
print(df.head(), '\n', df.shape)
```

```python
# Prepare data matrix
X = df[['Years','Hits']]
Y = (df['Salary']>=500).astype(int) # predict if salary is higher than 500 (instead of the exact amount)
```

```python
# Train a classification tree
from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=1, criterion='gini') # create a DT of depth 1
clf = clf.fit(X, Y)
```

```python
# Plot the tree
plt.figure(figsize=(10, 4))
tree.plot_tree(clf, feature_names=X.columns, filled=True);

# Save the plot as a PNG file
plt.savefig("classification_tree.png", dpi=300)
```

```python
# Root tree labels
Y.groupby(Y).count()
```

```python
# Proportion of each class in the root node
p_0 = 3/5
p_1 = 2/5
print('p_0 =', p_0, ', p_1 =', p_1)
```

```python
# Calculating Gini impurity (equivalent formula)
gini = p_0*(1-p_0) + p_1*(1-p_1)
print('Gini impurity:', gini)
```

```python
# Calculating Gini impurity
gini = 1 - p_0**2 - p_1**2
print('Gini impurity:', gini)
```
